**Ejercicio 1**

```python
# Mezcla de fertilizantes
import numpy as np
from scipy.optimize import minimize

# Datos del problema
# Ganancia por tonelada
g1 = 180   # Fertilizante F1
g2 = 220   # Fertilizante F2

# Disponibilidad de compuestos
A_disp = 120
B_disp = 80
C_disp = 150

# Función objetivo
# Como scipy.minimize minimiza, se usa el negativo para maximizar
# x[0] = x1 (toneladas de F1)
# x[1] = x2 (toneladas de F2)
def ganancia_negativa(x):
    x1, x2 = x[0], x[1]
    return -(g1 * x1 + g2 * x2)

# Restricciones de desigualdad
def restriccion_A(x):
    x1, x2 = x[0], x[1]
    return A_disp - (3*x1 + 4*x2)

def restriccion_B(x):
    x1, x2 = x[0], x[1]
    return B_disp - (2*x1 + x2)

def restriccion_C(x):
    x1, x2 = x[0], x[1]
    return C_disp - (5*x1 + 3*x2)

# Configuración inicial
x0 = [0, 0]

# Restricciones en formato Scipy
restricciones = [
    {'type': 'ineq', 'fun': restriccion_A},
    {'type': 'ineq', 'fun': restriccion_B},
    {'type': 'ineq', 'fun': restriccion_C}
]

# Límites de no negatividad
limites = [(0, None), (0, None)]

# Resolver el problema
resultado = minimize(
    ganancia_negativa,
    x0,
    method='SLSQP',
    constraints=restricciones,
    bounds=limites
)

# Resultado
x1_opt, x2_opt = resultado.x
ganancia_optima = -resultado.fun

print("Mezcla de fertilizantes:")
print("Toneladas óptimas de F1:", x1_opt)
print("Toneladas óptimas de F2:", x2_opt)
print("Ganancia máxima:", ganancia_optima)

# Verificación de restricciones
print("\nVerificación:")
print("Compuesto A usado:", 3*x1_opt + 4*x2_opt)
print("Compuesto B usado:", 2*x1_opt + x2_opt)
print("Compuesto C usado:", 5*x1_opt + 3*x2_opt)
```

**Ejercicio 2**

```python
# Produccion semanal de una fabrica
import numpy as np

# Datos del problema
# S_n = numero de sillas en la semana n
# M_n = numero de mesas en la semana n
# Condiciones iniciales
S0 = 200
M0 = 80

# Funciones de recurrencia
def sillas_sig(Sn, Mn):
    return 0.6 * Sn + 0.2 * Mn + 40

def mesas_sig(Sn, Mn):
    return 0.1 * Sn + 0.5 * Mn + 20

# Calculo de la semana 1
S1 = sillas_sig(S0, M0)
M1 = mesas_sig(S0, M0)

# Calculo de la semana 2
S2 = sillas_sig(S1, M1)
M2 = mesas_sig(S1, M1)

# Sistema para el equilibrio:
# S = 0.6S + 0.2M + 40
# M = 0.1S + 0.5M + 20
# Reordenando:
# 0.4S - 0.2M = 40
# -0.1S + 0.5M = 20

A = np.array([
    [0.4, -0.2],
    [-0.1, 0.5]
])

b = np.array([40, 20])

# Resolver el sistema de equilibrio
S_eq, M_eq = np.linalg.solve(A, b)

# Resultados
print("Produccion semanal de una fabrica:")
print("Semana 1:")
print("Sillas:", S1)
print("Mesas:", M1)

print("\nSemana 2:")
print("Sillas:", S2)
print("Mesas:", M2)

print("\nPunto de equilibrio:")
print("Sillas de equilibrio:", S_eq)
print("Mesas de equilibrio:", M_eq)

# Verificacion del equilibrio
S_ver = 0.6 * S_eq + 0.2 * M_eq + 40
M_ver = 0.1 * S_eq + 0.5 * M_eq + 20

print("\nVerificacion:")
print("S que sale de la recurrencia:", S_ver)
print("M que sale de la recurrencia:", M_ver)
```

**Ejercicio 3**

```python
# Epidemia en una ciudad
import numpy as np

# Datos del problema
# I(t) = numero de infectados
# S(t) = numero de susceptibles
# Poblacion total
N = 10000

# Parametros del modelo
beta = 0.00003
gamma = 0.1

# Condicion inicial
I0 = 50

# Funcion del modelo
# dI/dt = beta * S * I - gamma * I
# como S = N - I, entonces:
# dI/dt = beta * (N - I) * I - gamma * I
def modelo_epidemia(I):
    return beta * (N - I) * I - gamma * I

# Calculo de los puntos de equilibrio
# Se resuelve:
# beta * (N - I) * I - gamma * I = 0
# I * (beta * (N - I) - gamma) = 0
# Por tanto:
# I = 0
# o
# beta * (N - I) - gamma = 0

I_eq1 = 0
I_eq2 = N - (gamma / beta)

# Evaluacion del modelo en los equilibrios
f_eq1 = modelo_epidemia(I_eq1)
f_eq2 = modelo_epidemia(I_eq2)

# Resultados
print("Epidemia en una ciudad:")
print("Poblacion total:", N)
print("Beta:", beta)
print("Gamma:", gamma)
print("Condicion inicial I(0):", I0)

print("\nPuntos de equilibrio:")
print("Equilibrio 1 (sin infectados):", I_eq1)
print("Equilibrio 2 (endemia):", I_eq2)

# Verificacion
print("\nVerificacion:")
print("Modelo en I = 0:", f_eq1)
print("Modelo en I = equilibrio 2:", f_eq2)

# Interpretacion
print("\nInterpretacion:")
print("I = 0 significa que la enfermedad desaparece.")
print("I =", I_eq2, "significa un nivel constante de infectados.")
```

**Ejercicio 4**

```python
# Modelo logistico de pesca
import numpy as np

# Datos del problema
# Pn = poblacion de peces al final del ano n
P0 = 400   # poblacion inicial
r = 0.4    # tasa de crecimiento intrinseca
K = 1000   # capacidad maxima
H = 80     # peces extraidos por pesca

# Funcion de recurrencia
def poblacion_sig(Pn):
    return Pn + r * Pn * (1 - Pn / K) - H

# Calculo de los primeros anos
P1 = poblacion_sig(P0)
P2 = poblacion_sig(P1)

# Ecuacion de equilibrio:
# P = P + rP(1 - P/K) - H
# rP(1 - P/K) - H = 0
# 0.4P(1 - P/1000) - 80 = 0
# P^2 - 1000P + 200000 = 0

A = 1
B = -1000
C = 200000

discriminante = B**2 - 4*A*C

P_eq1 = (-B + np.sqrt(discriminante)) / (2*A)
P_eq2 = (-B - np.sqrt(discriminante)) / (2*A)

# Resultados
print("Modelo logistico de pesca:")
print("Poblacion inicial:", P0)

print("\nAno 1:")
print("Poblacion:", P1)

print("\nAno 2:")
print("Poblacion:", P2)

print("\nPuntos de equilibrio:")
print("Equilibrio 1:", P_eq1)
print("Equilibrio 2:", P_eq2)

# Verificacion de los equilibrios
P_ver1 = poblacion_sig(P_eq1)
P_ver2 = poblacion_sig(P_eq2)

print("\nVerificacion:")
print("Valor al aplicar la recurrencia en equilibrio 1:", P_ver1)
print("Valor al aplicar la recurrencia en equilibrio 2:", P_ver2)
```
